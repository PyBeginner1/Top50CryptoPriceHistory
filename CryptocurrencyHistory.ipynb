```python
import pandas as pd
import datetime as dt
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import pandas_datareader.data as web
```

#Solana

```python
sol_data = pd.read_csv('/content/Crypto/Solana.csv')
```

```python
sol_data
```

```python
sol_data.drop('SNo', axis = 1, inplace = True)
```

```python
sol_data
```

```python
sol_data.describe()
```

```python
fig = px.area(data_frame = sol_data, y = 'Price', x = 'Date', title = 'Solana Daily Price')

fig.show()
```

```python
fig = px.area(data_frame = sol_data, y = 'Price', x = 'Date', title = 'Solana Daily Price')

fig.update_xaxes(
    title_text = 'Date',
    rangeslider_visible = True,
    rangeselector = dict(
        buttons = list([
            dict(count = 1, label = '1M', step = 'month', stepmode = 'backward'),
            dict(count = 6, label = '6M', step = 'month', stepmode = 'backward'),
            dict(count = 1, label = 'YTD', step = 'year', stepmode = 'todate'),
            dict(count = 1, label = '1Y', step = 'year', stepmode = 'backward'),
            dict(step='all')
        ])
    )
)

fig.update_yaxes(
    title_text = 'Price in USD',
    ticksuffix = '$'
)

fig.show()
```

```python
fig = go.Figure(data = [go.Candlestick(x = sol_data['Date'],
                                       open = sol_data['Open'],
                                       high = sol_data['High'],
                                       low = sol_data['Low'],
                                       close = sol_data['Price'])])

fig.update_xaxes(
    title_text = 'Date',
    rangeselector = dict(
        buttons = list([
                        dict(count = 1, label = '1M', step = 'month', stepmode = 'backward'),
                        dict(count = 6, label = '6M', step = 'month', stepmode = 'backward'),
                        dict(count = 1, label = 'YTD', step = 'year', stepmode = 'todate'),
                        dict(count = 1, label = '1Y', step = 'year', stepmode = 'backward'),
                        dict(step='all')
        ])
    )
)

fig.update_yaxes(
    title_text = 'Price in USD',
    ticksuffix = '$'
)


fig.show()
```

```python
fig = px.line(data_frame = sol_data,
              x = 'Date',
              y = 'Change %')

fig.update_xaxes(
    title_text = 'Date',
    rangeselector = dict(
        buttons = list([
                        dict(count = 1, label = '1M', step = 'month', stepmode = 'backward'),
                        dict(count = 6, label = '6M', step = 'month', stepmode = 'backward'),
                        dict(count = 1, label = 'YTD', step = 'year', stepmode = 'todate'),
                        dict(count = 1, label = '1Y', step = 'year', stepmode = 'backward'),
                        dict(step='all')
        ])
    )
)

fig.update_yaxes(
    title_text = 'Price in USD',
    ticksuffix = '$'
)

fig.show()
```

#Top 50 Cryptocurrencies

```python
data = pd.read_csv('/content/Crypto/ALL_COMBINED.csv')
```

```python
data
```

```python
data.describe()
```

```python
fig = px.area(data_frame = data,
              x = 'Date',
              y = 'Price',
              line_group = 'Currency_Name',
              color = 'Currency_Name',
              title = 'Price of Top 50 Cryptocurrencies')

fig.update_xaxes(
    title_text = 'Date',
    rangeselector = dict(
        buttons = list([
                        dict(count = 1, label = '1M', step = 'month', stepmode = 'backward'),
                        dict(count = 6, label = '6M', step = 'month', stepmode = 'backward'),
                        dict(count = 1, label = 'YTD', step = 'year', stepmode = 'todate'),
                        dict(count = 1, label = '1Y', step = 'year', stepmode = 'backward'),
                        dict(step='all')
        ])
    )
)

fig.update_yaxes(title_text = 'Price in USD', ticksuffix = '$')

fig.show()
```

```python
fig =px.area(data_frame = data,
                 y  = "Change %" , 
                 x = "Date" , 
                 line_group="Currency_Name", 
                 color = "Currency_Name",
                 color_discrete_sequence=px.colors.qualitative.Alphabet,
              title = 'Daily Percentage Change of all 50 Cryptocurrencies')

fig.update_xaxes(
    title_text = 'Date',
    rangeselector = dict(
        buttons = list([
                        dict(count = 1, label = '1M', step = 'month', stepmode = 'backward'),
                        dict(count = 6, label = '6M', step = 'month', stepmode = 'backward'),
                        dict(count = 1, label = 'YTD', step = 'year', stepmode = 'todate'),
                        dict(count = 1, label = '1Y', step = 'year', stepmode = 'backward'),
                        dict(step='all')
        ])
    )
)

fig.update_yaxes(title_text = 'Price in USD', ticksuffix = '$')

fig.show()
```
